```python
import cv2
import pywt
import matplotlib.pyplot as plt

def decompose_image(image, wavelet='haar', levels=5):
    # Melakukan dekomposisi citra menggunakan wavelet
    coeffs = pywt.wavedec2(image, wavelet, level=levels)
    return coeffs

def display_coefficient(coefficients, level):
    # Menampilkan koefisien hasil dekomposisi
    fig, axs = plt.subplots(1, 3, figsize=(12, 4))
    titles = ['Approximation', 'Horizontal detail', 'Vertical detail', 'Diagonal detail'][:len(coefficients)]

    for i, (coeff, title) in enumerate(zip(coefficients, titles)):
        axs[i].imshow(coeff, cmap='gray')
        axs[i].set_title(f'{title} (Level {level})')
        axs[i].axis('off')

    plt.tight_layout()
    plt.show()

def main():
    # Memuat citra grayscale
    image = cv2.imread('IoriYagami.jpg', cv2.IMREAD_GRAYSCALE)

    # Melakukan dekomposisi citra
    coeffs = decompose_image(image)

    # Menampilkan koefisien untuk level yang berbeda
    for level in range(1, len(coeffs)):
        display_coefficient(coeffs[level], level)

    # Menampilkan citra asli
    plt.figure(figsize=(4, 4))
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    plt.show()

if __name__ == "__main__":
    main()
```
